```python
import matplotlib.pyplot as plt
```

```python
x = [10, 12, 14, 16]
y = [21, 23, 25, 27]
```

```python
plt.plot(x,y)
```

```python
z = [5, 9, 15, 10]
```

```python
# Two line graph in same plot
plt.plot(z,y)
plt.plot(z,x)
```

```python
# Two different graph types
f, ax = plt.subplots(2)
f.suptitle("X AXIS")
ax[0].plot(x,y)
ax[1].scatter(x,y)
```

```python
#Changing Scatter Sign as '*'
plt.plot(x,y, "*")
```

```python
#Changing Line Color to red (r)
plt.plot(x,y, "r")
```

```python
#Changing Scatter Color to yellow(y)
plt.plot(x,y, "*y")
```

```python
# Labeling Graph
plt.title("My Graph")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.plot(x,y)
```
